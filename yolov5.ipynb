```python
import os
import csv

# YOLO 라벨 디렉토리
label_dir = '/content/drive/MyDrive/deep/exp/labels'

# CSV 파일 저장 경로
csv_output_path = '/content/drive/MyDrive/deep/detection_results.csv'

# 클래스 이름 정의 (YOLOv5의 COCO 데이터셋 클래스 예제)
classes = [
    "person", "bicycle", "car", "motorbike", "aeroplane", "bus", "train", "truck", "boat", "traffic light",
    "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat", "dog", "horse", "sheep", "cow",
    "elephant", "bear", "zebra", "giraffe", "backpack", "umbrella", "handbag", "tie", "suitcase", "frisbee",
    "skis", "snowboard", "sports ball", "kite", "baseball bat", "baseball glove", "skateboard", "surfboard",
    "tennis racket", "bottle", "wine glass", "cup", "fork", "knife", "spoon", "bowl", "banana", "apple",
    "sandwich", "orange", "broccoli", "carrot", "hot dog", "pizza", "donut", "cake", "chair", "sofa",
    "pottedplant", "bed", "diningtable", "toilet", "TV monitor", "laptop", "mouse", "remote", "keyboard",
    "cell phone", "microwave", "oven", "toaster", "sink", "refrigerator", "book", "clock", "vase",
    "scissors", "teddy bear", "hair drier", "toothbrush"
]

# CSV 파일 작성
with open(csv_output_path, mode='w', newline='') as csv_file:
    writer = csv.writer(csv_file)

    # CSV 헤더 작성
    writer.writerow(["Image", "Class", "Bounding Box (x_center, y_center, width, height)"])

    if os.path.exists(label_dir):
        label_files = os.listdir(label_dir)
        print(f"총 라벨 파일 수: {len(label_files)}")

        for label_file in label_files:
            label_path = os.path.join(label_dir, label_file)
            with open(label_path, 'r') as f:
                for line in f.readlines():
                    parts = line.strip().split()
                    class_id = int(parts[0])  # 클래스 ID
                    bbox = parts[1:]          # 바운딩 박스 정보 (x_center, y_center, width, height)

                    # CSV 행 추가
                    writer.writerow([label_file.replace('.txt', '.jpg'), classes[class_id], ", ".join(bbox)])
        print(f"CSV 파일 저장 완료: {csv_output_path}")
    else:
        print(f"라벨 디렉토리가 존재하지 않습니다: {label_dir}")
```

```python
import os
from collections import Counter

# YOLO 라벨 디렉토리
label_dir = '/content/drive/MyDrive/deep/exp/labels'

# 클래스 이름 정의 (YOLOv5의 COCO 데이터셋 클래스 예제)
classes = [
    "person", "bicycle", "car", "motorbike", "aeroplane", "bus", "train", "truck", "boat", "traffic light",
    "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat", "dog", "horse", "sheep", "cow",
    "elephant", "bear", "zebra", "giraffe", "backpack", "umbrella", "handbag", "tie", "suitcase", "frisbee",
    "skis", "snowboard", "sports ball", "kite", "baseball bat", "baseball glove", "skateboard", "surfboard",
    "tennis racket", "bottle", "wine glass", "cup", "fork", "knife", "spoon", "bowl", "banana", "apple",
    "sandwich", "orange", "broccoli", "carrot", "hot dog", "pizza", "donut", "cake", "chair", "sofa",
    "pottedplant", "bed", "diningtable", "toilet", "TV monitor", "laptop", "mouse", "remote", "keyboard",
    "cell phone", "microwave", "oven", "toaster", "sink", "refrigerator", "book", "clock", "vase",
    "scissors", "teddy bear", "hair drier", "toothbrush"
]

# 클래스별 탐지 카운트 저장
class_counts = Counter()

# 라벨 파일 읽기
if os.path.exists(label_dir):
    label_files = os.listdir(label_dir)
    print(f"총 라벨 파일 수: {len(label_files)}")

    for label_file in label_files:
        label_path = os.path.join(label_dir, label_file)
        with open(label_path, 'r') as f:
            for line in f.readlines():
                class_id = int(line.strip().split()[0])  # 클래스 ID 추출
                class_counts[class_id] += 1  # 클래스 ID에 해당하는 카운트 증가
else:
    print(f"라벨 디렉토리가 존재하지 않습니다: {label_dir}")

# 결과 출력
print("\n전체 객체별 탐지 결과:")
for class_id, count in class_counts.items():
    print(f"{classes[class_id]}: {count}개")
```
